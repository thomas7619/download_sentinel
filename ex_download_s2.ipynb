```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
#---------
bands=["B02","B03","B04"]

startmonth = "202109"
endmonth = "202208"
pas = "month"   # ou 10, 20, etc.
#pas = 12
input_path = 'rennes_se.geojson'
max_cloud_cover =99
#---------

start = datetime.strptime(startmonth, "%Y%m")
end = datetime.strptime(endmonth, "%Y%m")
output_folder = './outputs/rennes_se/alldata_months/'

current = start
while current <= end:
    startday = current.strftime("%Y%m%d")
    if pas == "month":
        next_date = current + relativedelta(months=1)
    else:
        next_date = current + timedelta(days=int(pas))

    endday = (next_date - timedelta(days=1)).strftime("%Y%m%d")
    output_path_data='%s/%s'%(output_folder,startday)
    print(f"{startday} → {endday}")
    S2.download_S2(input_path,output_path_data,startday,endday,bands=bands,visu=True,synthesis=True,max_cloud_cover=max_cloud_cover)
    current = next_date
    
```
