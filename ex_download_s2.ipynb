# 2) Example with the download of a synthesis over June 2024

The first connection to the server may be unstable. In this case, try several times

## 1) Downloading a synthesis over a given period

```python
bands=["B01","B02","B03","B04","B05","B06","B07","B8A", "B09", "B11", "B12"]
input_path = 'copacabana_ipanema.geojson'
output_folder = './outputs/copacabana_ipanema_synthesis_2024_06/'
startday = "20240601"
endday = "20240615"
synthesis = True # To download a ynthesis over the time period
S2.download_S2(input_path,output_folder,startday,endday,bands=bands,visu=True,synthesis=synthesis)
```
